```python
from PIL import Image, ExifTags
import os
```

```python
def get_exif_data(image):
    """Extrahiert EXIF-Daten (falls vorhanden) und macht sie lesbar."""
    exif_data = {}
    try:
        exif = image._getexif()
        if exif is not None:
            for tag, value in exif.items():
                decoded = ExifTags.TAGS.get(tag, tag)
                exif_data[decoded] = value
    except Exception:
        pass
    return exif_data

def analyze_image(path):
    """Analysiert eine einzelne JPEG-Datei."""
    try:
        with Image.open(path) as img:
            file_size = os.path.getsize(path)

            print(f"\n📷 Datei: {path}")
            print(f"   Größe (Bytes): {file_size}")
            print(f"   Format: {img.format}")
            print(f"   Modus: {img.mode}")
            print(f"   Auflösung: {img.size[0]} x {img.size[1]} Pixel")

            # DPI (falls vorhanden)
            if hasattr(img, "info") and "dpi" in img.info:
                print(f"   DPI: {img.info['dpi']}")

            # EXIF Daten
            exif = get_exif_data(img)
            if exif:
                print("   EXIF-Daten:")
                for k, v in exif.items():
                    print(f"      {k}: {v}")
            else:
                print("   Keine EXIF-Daten gefunden.")

    except Exception as e:
        print(f"❌ Fehler bei {path}: {e}")

def scan_folder(folder):
    """Durchsucht einen Ordner nach JPEG-Dateien."""
    for root, _, files in os.walk(folder):
        for file in files:
            if file.lower().endswith((".jpg", ".jpeg")):
                analyze_image(os.path.join(root, file))

if __name__ == "__main__":
    folder_path = "recoating"
    scan_folder(folder_path)
```
